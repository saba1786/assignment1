```python
driver = webdriver.Chrome()
```

```python
url = "https://www.billboard.com/"
driver.get(url)
```

```python
song_names = driver.find_elements(By.ID,'title-of-a-story')
song_names
```

```python
#extracting the text  from these tags 
Song_Name=[]
for i in song_names:
    Song_Name.append(i.text)
Song_Name[:100]
```

```python

# Remove empty strings, commas, and quotes
song_name = [item.replace(",", "").replace("'", "").replace('"', '').strip() for item in Song_Name if item.strip()]

# Print cleaned data
print(song_name)
```

```python
artist_name=driver.find_elements(By.XPATH,"//li[@class='o-chart-results-list__item // lrv-u-flex-grow-1 lrv-u-flex lrv-u-flex-direction-column lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light lrv-u-padding-l-050 lrv-u-padding-l-1@mobile-max']")
artist_name
```

```python
#extracting the text  from these tags 
Artist_Name=[]
for j in artist_name:
    Artist_Name.append(j.text)
Artist_Name[:100]
```

```python
# Remove newline characters (\n)
artist_name = [item.replace("\n", "") for item in artist_name]

# Print cleaned data
print(artist_name)
```
