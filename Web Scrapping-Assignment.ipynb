```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Question 1

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
bs = BeautifulSoup(page.content, 'html.parser')
titles = bs.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

```python
df=pd.DataFrame({'Titles':titles})
df
```

# Question 2

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
data = []
for div in soup.find_all('div', class_='desc-sec'):
    president_name = div.h3.text.strip()
    president_title = div.h5.text.strip()
    data.append({'President Name': president_name, 'President Title': president_title})
df = pd.DataFrame(data)
print(df)
```

# Question 5

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
headlines = []
for headline in soup.select('div.Card-titleContainer'):
    headlines.append(headline.text.strip())

df = pd.DataFrame({'Headline': headlines})
print(df)
```

```python
times = []
for i in soup.select('time.LatestNews-timestamp'):
    times.append(i.text.strip())

df = pd.DataFrame({'Time': times})
print(df)
```

```python
df=pd.DataFrame({'Headline':headlines,'Time':times})
df
```

# Question 6

```python
page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
titles = []
for title in soup.find_all('h2', class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    titles.append(title.text)

df = pd.DataFrame({'Paper Title': titles})
print(df)
```

```python
authors = []
for author in soup.find_all('span', class_='sc-1w3fpd7-0 dnCnAO'):
    authors.append(author.text)

df = pd.DataFrame({'Author': authors})
print(df)
```

```python
dates = []
for date in soup.find_all('span', class_='sc-1thf9ly-2 dvggWt'):
    dates.append(date.text)
df=pd.DataFrame({'Dates':dates})
print(df)
```

```python
import pandas as pd
df=pd.DataFrame({'Paper Title':titles,'Author':authors,'Dates':dates})
df
```

```python
price=soup.find('span',class_="double-line-ellipsis")
price.text.replace('₹','')
```

```python
titles=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
titles
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.replace('₹',''))
price
```

```python
images=[]
for i in soup.find_all('img',class_="no-img"):
    images.append(i['data-src'])
images
```

```python
import pandas as pd 
df=pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images_url':images})
df
```
